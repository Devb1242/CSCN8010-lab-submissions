```python
# Importing necessary Packages
from tensorflow.keras.datasets import fashion_mnist
import matplotlib.pyplot as plt
import numpy as np

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from sklearn.metrics import confusion_matrix ,classification_report
```

```python
# Loading the Fashion MNIST dataset
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
```

```python
# Exploring the dataset
print("Training Images Set Shape:", train_images.shape)
print("Number of Training Images Samples:", len(train_images))
print("Image dimensions:", train_images[0].shape)

print("\nTest Images Set Shape:", test_images.shape)
print("Number of Test Images Samples:", len(test_images))
print("Image dimensions:", test_images[0].shape)
```

```python
# Displaying some images from the dataset
fig, axes = plt.subplots(5, 5, figsize=(10, 6))
axes = axes.ravel()

for i in np.arange(25):
    axes[i].imshow(train_images[i], cmap='gray')
    axes[i].axis('off')

plt.tight_layout()
plt.show()
```

* Grayscale photos of various fashion items are included in the fMNIST dataset, which is an adaptation of the popular MNIST dataset.
* The dataset contains grayscale images of 10 different categories of fashion items.
* It contains **60,000** sample images of training and **10,000** sample images of testing data and each images occupy a space of **28x28** pixels.

#### 2. Train a dense neural network using Keras, try out at least three hyperparameters configurations

```python
# Normalizing the image's pixels value to the range of [0,1]
train_images = train_images.astype("float32") / 255.0
test_images = test_images.astype("float32") / 255.0

# Now, flattening the images for the dense neural network
train_images = train_images.reshape((60000, 28*28))
test_images = test_images.reshape((10000, 28*28))
```

* Model-1
    * Training the 1st model by taking **128** as an input.

```python
DNN_model1 = Sequential([
        Dense(128, activation='relu'),
        Flatten(input_shape=(train_images.shape[1],)),
        Dense(512, activation='relu'),
        Dense(64, activation='relu'),
        Dense(10, activation='softmax')
    ])
DNN_model1.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
print("Model-1")
DNN_model1.fit(train_images,train_labels, epochs=20)
```

* Model-2
    * Training the 2nd model by taking **512** as an input

```python
DNN_model2 = Sequential([
        Dense(512, activation='relu'),
        Flatten(input_shape=(train_images.shape[1],)),
        Dense(64, activation='relu'),
        Dense(32, activation='relu'),
        Dense(10, activation='softmax')
    ])
DNN_model2.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
print("Model-2:")
history2 = DNN_model2.fit(train_images,train_labels, epochs=15)
```

* Model-3
    * Training 3rd model by taking **32** as an input

```python
DNN_model3 = Sequential([
        Dense(32, activation='relu'),
        Flatten(input_shape=(train_images.shape[1],)),
        Dense(32, activation='relu'),
        Dense(10, activation='softmax')
    ])
DNN_model3.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
print("Model-3:")
DNN_model3.fit(train_images,train_labels, epochs=10)
```

#### 3. Evaluate its accuracy for each of the configurations. Choose the best configuration, and: Report its accuracy, confusion metric, and specific examples of incorrect classification, and drew some conclusions from this exploration

* Checking the accuracy for each of the configurations.

```python
def prediction_model(model, x_image):
    y_pred = np.argmax(model.predict(x_image.reshape((-1, 28 * 28))), axis=1)
    return y_pred

def model_evaluation(y_pred , y_test):
    report = classification_report(y_test, y_pred,output_dict=True)
    return (report)
```

```python
y_pred_1 = prediction_model(DNN_model1, test_images)
print("Model-1:",pd.DataFrame(model_evaluation(y_pred_1, test_labels)).T)
```

```python
y_pred_2 = prediction_model(DNN_model2, test_images)
print("Model-2:",pd.DataFrame(model_evaluation(y_pred_2, test_labels)).T)
```

```python
y_pred_3 = prediction_model(DNN_model3, test_images)
print("Model-3",pd.DataFrame(model_evaluation(y_pred_3, test_labels)).T)
```

* From the above models accuracy **Model-1** has the best accuracy among other model.
* Its accuracy is **89%**.

```python
confusion_matrix(y_pred_1, test_labels)
```
